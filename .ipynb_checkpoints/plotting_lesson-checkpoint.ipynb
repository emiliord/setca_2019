## Plotting Lesson

```python
import numpy
distances = numpy.genfromtxt(fname='distance_data_headers.csv', delimiter=',', dtype='unicode')
print(distances)
```

```python
headers = distances[0]
data = distances[1:]
data = data.astype(numpy.float)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(data[:,1])
```

```python
plt.plot(data[:,0],data[:,1])
```

```python
sample = headers[1]
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
fig_1 = plt.plot(data[:,0], data[:,1], label=sample)
plt.legend()

plt.savefig(F'{sample}.png')
```

```python
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
plt.plot(data[:,0], data[:,1], label=headers[1])
plt.plot(data[:,0], data[:,2], label=headers[2])
plt.legend()
```

```python
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
plt.plot(data[:,0], data[:,2], label=headers[2])
plt.plot(data[:,0], data[:,1], label=headers[1])
plt.legend()
```

```python
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
for col in range(1, len(headers)):
    fig = plt.plot(data[:,0], data[:,col], label=headers[col])
plt.legend()
```

```python
for col in range(1, len(headers)):
    plt.figure()
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstroms)')
    fig = plt.plot(data[:,0], data[:,col], label=headers[col])
    plt.legend()
    plt.savefig(F'{headers[col]}.png', dpi=300)
```
